```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
import os  # perbaikan dari 'oss'

# URL gambar dan nama file lokal
url = "https://picsum.photos/400/300"
filename = "unpam.jpg"  # simpan di direktori saat ini

# Unduh gambar jika belum ada
if not os.path.exists(filename):
    urllib.request.urlretrieve(url, filename)

# Membaca citra dalam mode grayscale
image = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

# Cek apakah gambar berhasil dibaca
if image is None:
    raise ValueError("Gambar gagal dibaca. Periksa URL atau path file.")

# Melakukan operasi negatif citra
negative_image = 255 - image

# Menampilkan citra asli dan negatif
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Citra Asli')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Negatif Citra')
plt.imshow(negative_image, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
import os
# Download gambar sample
filename = "gw.jpg"
if not os.path.exists(filename):
 urllib.request.urlretrieve(url, filename)
# Membaca citra dalam mode grayscale
image = cv2.imread(filename, 0)
# Menentukan konstanta skala (c)
c = 255 / np.log(1 + np.max(image))
# Menerapkan transformasi logaritma pada citra
log_transform = c * (np.log(1 + image.astype(np.float64)))
# Normalisasi hasil transformasi agar sesuai dengan format citra 8-bit
log_transform = np.array(log_transform, dtype=np.uint8)
# Menampilkan citra asli dan hasil transformasi logaritma
plt.figure(figsize=(10, 5))
# Menampilkan citra asli
plt.subplot(1, 2, 1)
plt.title('Citra Asli')
plt.imshow(image, cmap='gray')
plt.axis('off') # Menghilangkan axis
# Menampilkan hasil transformasi logaritma
plt.subplot(1, 2, 2)
plt.title('Transformasi Logaritma')
plt.imshow(log_transform, cmap='gray')
plt.axis('off') # Menghilangkan axis
# Menampilkan plot
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import urllib.request
import os
# Download gambar sample
url = "https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg"
filename = "gw.jpg"
if not os.path.exists(filename):
 urllib.request.urlretrieve(url, filename)
# Membaca citra dalam mode grayscale
image = cv2.imread(filename, 0)
# Fungsi untuk melakukan Gamma Correction
def gamma_correction(image, gamma):
 # Konversi citra ke tipe float32 untuk menghindari clipping
 image_float = image.astype(np.float32) / 255.0

# Melakukan transformasi pangkat-law (gamma correction)
 corrected_image = np.power(image_float, gamma)
 # Mengembalikan ke rentang 0-255 dan konversi ke tipe uint8
 corrected_image = np.uint8(corrected_image * 255)
 return corrected_image
# Nilai gamma
gamma = 2.2 # Ubah nilai gamma untuk melihat perbedaan
corrected_image = gamma_correction(image, gamma)
# Tampilkan citra asli dan citra setelah gamma correction
plt.figure(figsize=(10, 5))
# Citra Asli
plt.subplot(1, 2, 1)
plt.title('Citra Asli')
plt.imshow(image, cmap='gray')
plt.axis('off')
# Citra setelah Gamma Correction
plt.subplot(1, 2, 2)
plt.title('Citra Setelah Gamma Correction (gamma = {})'.format(gamma))
plt.imshow(corrected_image, cmap='gray')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import urllib.request
import os

# Download gambar sample
filename = "gw.jpg"
if not os.path.exists(filename):
 urllib.request.urlretrieve(url, filename)
# Membaca citra
image = cv2.imread(filename)
# Menerapkan filter rata-rata
average_blur = cv2.blur(image, (5, 5))
# Menerapkan filter Gaussian
gaussian_blur = cv2.GaussianBlur(image, (5, 5), 0)
# Menampilkan citra asli dan hasil perataan
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title('Citra Asli')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(1, 3, 2)
plt.title('Perataan (Rata-rata)')
plt.imshow(cv2.cvtColor(average_blur, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(1, 3, 3)
plt.title('Perataan (Gaussian)')
plt.imshow(cv2.cvtColor(gaussian_blur, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

try:
    # Membaca citra
    image = cv2.imread('gw.jpg', cv2.IMREAD_GRAYSCALE)

    if image is None:
        # Buat gambar dummy jika file tidak ditemukan
        print("File tidak ditemukan, menggunakan gambar dummy...")
        image = np.random.randint(0, 255, (300, 400), dtype=np.uint8)

    # Membuat Filter Laplacian untuk sharpening
    laplacian_filter = np.array([[0, -1, 0],
                                 [-1, 5, -1],
                                 [0, -1, 0]])

    # Menerapkan filter Laplacian untuk penajaman
    sharpened_image = cv2.filter2D(image, -1, laplacian_filter)

    # Menampilkan Hasil
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.title("Citra Asli")
    plt.imshow(image, cmap='gray')
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.title("Citra Setelah Penajaman (Laplacian Filter)")
    plt.imshow(sharpened_image, cmap='gray')
    plt.axis("off")

    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"Terjadi error: {e}")
```

```python
import cv2
import numpy as np
try:
 # Membaca citra
 image = cv2.imread('sepul1.jfif', 0)

 # Cek jika gambar berhasil dibaca
 if image is None:
 print("File 'images.jpg' tidak ditemukan. Membuat gambar dummy...")
 # Buat gambar dummy dengan noise untuk demonstrasi
 image = np.random.randint(0, 255, (300, 400), dtype=np.uint8)
 # Tambahkan noise salt & pepper
noise = np.random.random((300, 400))
 image[noise < 0.05] = 0 # Noise hitam
 image[noise > 0.95] = 255 # Noise putih

 # Menerapkan median filtering
 filtered_image = cv2.medianBlur(image, 5) # 5 adalah ukuran jendela (kernel)

 # Tampilkan citra asli dan citra setelah median filtering
 plt.figure(figsize=(12, 6))

 plt.subplot(1, 2, 1)
 plt.title('Citra Asli')
 plt.imshow(image, cmap='gray')
 plt.axis('off')

 plt.subplot(1, 2, 2)
 plt.title('Citra Setelah Median Filtering\n(Kernel size: 5)')
 plt.imshow(filtered_image, cmap='gray')
 plt.axis('off')
 plt.tight_layout()
 plt.show()

 print(f"Ukuran citra: {image.shape}")
 print("Median filter berhasil diterapkan!")
except Exception as e:
print(f"Error: {e}")
```
